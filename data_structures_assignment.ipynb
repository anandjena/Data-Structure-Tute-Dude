### Task-3(Product Pricing (Dictionaries))

```python
#1- Create a dictionary price_dict where keys are product names and values are prices (integers or floats). Include at least 6 entries.
price_dict = {
    "Laptop": 45000,
    "Smartphone": 20000,
    "Headphones": 5000,
    "Camera": 15000,
    "Smartwatch": 8000,
    "Tablet": 12000,
    "Printer": 7000
}
print(price_dict)
#2-
# Add a new product with price to price_dict.
price_dict['Soundbar'] = 15000
print(price_dict)
# Update the price of an existing product.
price_dict['Laptop'] = 65000
print(price_dict)
#Remove a product by name (handle the case when the product does not exist).
removed_product = price_dict.pop('camera',None)
if removed_product is not None:
    print("product removed successfully.")
else:
    print("Product not found.")
print(price_dict)

#Extra (optional): Print the product with both the maximum and minimum prices.
max_price_product = max(price_dict,key = price_dict.get)
print(f"max price product:{max_price_product} with price {price_dict[max_price_product]}")
min_price_procut = min(price_dict, key = price_dict.get)
print(f"min price product:{min_price_procut} with price {price_dict[min_price_procut]}")
```

## Task -4  Combined Operations

```python
#1-Using the products list and price_dict, create a list of tuples named catalog where each tuple is (product_name, price, category).

categories.append("Electrical Appliances")
categories.append("Sports Equipment")

catalog = [
    (value, price_dict[value], categories[index])
    for index, value in enumerate(product)
]
print(catalog)

#2- From catalog, create a new dictionary category_to_products that maps each category to a list of product names in that category.
category_to_products = {}

for name, price, category in catalog:
    if category not in category_to_products:
        category_to_products[category] = []
    
    category_to_products[category].append(name)

print(category_to_products)

#3- Print all products that belong to the category that has the maximum number of products.
max_count = 0
max_category = None

for category in category_to_products:
    count = len(category_to_products[category])
    
    if count > max_count:
        max_count = count
        max_category = category

print("Category with maximum products:", max_category)
print("Products:", category_to_products[max_category])
```
